```python
#load the data
import pandas as pd
base = pd.read_csv('dados_faltantes_exercicio1.csv')
base.info()
```

```python
#display the first five rows
base.head()
```

```python
#counting the missing values by column
base.isna().sum()
```

```python
#identifying the missing values of 'Idade'
base.loc[base.Idade.isnull(),'Idade']
```

```python
#identifying the missing values of 'Renda Anual'
base.loc[base['Renda Anual'].isnull(),'Renda Anual']
```

```python
#Fill the gaps in the 'Idade' column with the mean
base['Idade']=base['Idade'].fillna(base['Idade'].mean())
base.info()
```

```python
#Remove the rows with missings values in the column 'Renda Anual'
base = base.dropna(subset=['Renda Anual'])
```

```python
#display the data
display(base)
```

```python
#saving the data
base.to_csv('dados_limpos.csv')
```
