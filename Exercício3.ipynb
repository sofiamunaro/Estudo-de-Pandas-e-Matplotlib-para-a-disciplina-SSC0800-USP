```python
#load the data
import pandas as pd
base = pd.read_csv('clientes.csv')
base.info()
```

```python
#checking duplicates
base.duplicated()
```

```python
#removing duplicates
base.drop_duplicates()
base.info()
```

```python
#applying filter
base = base[base['Valor da Compra']>500]
base = base[base['Cidade']=='São Paulo']
display(base)
```

```python
#saving the data
base.to_csv('clientes_filtrados.csv')
```

```python
import matplotlib.pyplot as plt
```

```python
#histogram
plt.hist(base['Valor da Compra'],edgecolor='black')
plt.title('Valor da Compra')
```

```python
#boxplot
plt.boxplot(base['Valor da Compra'])
```
